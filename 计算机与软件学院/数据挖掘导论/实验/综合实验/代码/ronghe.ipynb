```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score, accuracy_score, mean_absolute_error, f1_score
```

```python
df_oof_lgb = pd.read_csv('lgb_oof.csv')
df_oof_xgb = pd.read_csv('xgb_oof.csv')
df_oof_ctb = pd.read_csv('ctb_oof.csv')
```

```python
df_oof_lgb.columns = ['SaleID', 'price', 'lgb_pred']
df_oof_xgb.columns = ['SaleID', 'price', 'xgb_pred']
df_oof_ctb.columns = ['SaleID', 'price', 'ctb_pred']
```

```python
del df_oof_lgb['price']
del df_oof_ctb['price']
```

```python
df_oof = df_oof_xgb.merge(df_oof_lgb, on='SaleID', how='left')\
    .merge(df_oof_ctb, on='SaleID', how='left')
# df_oof = pd.concat([df_oof,df_oof_ctb],axis=1,on='SaleID')
# df_oof = df_oof..merge(df_oof_ctb, on='SaleID', how='left')
```

```python
df_oof.head()
```

```python
min_mae = 9999
minA, minB, minC = 0, 0, 0
for a in range(100):
    for b in range(100-a):
        c = 100-a-b
        df_oof['pred'] = a/100*df_oof['ctb_pred']+b/100 * \
            df_oof['xgb_pred'] + c/100 * df_oof['lgb_pred']
        mae = mean_absolute_error(df_oof['price'], df_oof['pred'])
        if(mae < min_mae):
            minA, minB, minC = a, b, c
            min_mae = mae
            print(min_mae, minA, minB, minC)
```

```python
df_oof['pred'] = minA/100*df_oof['ctb_pred']+minB/100 * \
    df_oof['xgb_pred'] + minC/100 * df_oof['lgb_pred']
mae = mean_absolute_error(df_oof['price'], df_oof['pred'])
print('mae:', mae)
```
